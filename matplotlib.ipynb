```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x=[1,2,3,4,5,6,7,8]
y=[20,30,40,50,90,40,45,60]
```

```python
#API for plot
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('weather')



plt.plot(x,y,color='green', linewidth=5, linestyle='dashdot')
```

```python
#documentation for plot function formate string
plt.plot(x,y, 'rD')
```

```python
#individual argument for plot
plt.plot(x,y,color='blue',marker='D', linestyle='', markersize=20)
```

```python
#alpha property of plot API
plt.plot(x,y,color='blue',alpha=0.3)
```

```python
days=[1,2,3,4,5,6,7]
max_t=[30,40,50,60,70,80,60]
min_t=[15,20,25,10,16,28,29]
avg_t=[25,38,45,57,68,78,59]
```

```python
#create legend
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")


plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")

plt.plot(days,avg_t, label="Avg")
plt.legend(loc="best", shadow=True, fontsize="small")
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#barchart
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
#horizontal barchart

plt.yticks(ypos,company)

plt.title("US Tech Stocks")
plt.barh(xpos-0.2,revenue,  label="Revenue")
plt.barh(xpos+0.2,profit, label="Profit")
plt.legend()
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, bins=[80,100,125,200], rwidth=0.95, color='r')
```

```python
#API for hist function
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['blue','orange'], label=['men','women'])
plt.legend()

```

```python
#orientation property

blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['blue','orange'], label=['men','women'], orientation='horizontal')
plt.legend()
```
